## Part 3: Portfolio Demo Template

Here's a template you can customize for your capstone project.

````python
import os

portfolio_dir = '/tmp/portfolio_demo'
os.makedirs(portfolio_dir, exist_ok=True)

# Gradio Portfolio Template
gradio_template = '''"""Portfolio Demo - [Your Project Name]

A professional demo showcasing [brief description of your project].

Author: [Your Name]
GitHub: [Your GitHub URL]
LinkedIn: [Your LinkedIn URL]
"""

import gradio as gr
import time

# ============================================================================
# CONFIGURATION
# ============================================================================

PROJECT_TITLE = "ðŸš€ [Your Project Name]"
PROJECT_DESCRIPTION = """
[One-paragraph description of what your project does and why it matters]

**Key Features:**
- Feature 1: [Brief description]
- Feature 2: [Brief description]
- Feature 3: [Brief description]

**Technologies Used:** [List key technologies]
"""

# Golden examples that always work
EXAMPLES = [
    ["Example input 1 that showcases feature A"],
    ["Example input 2 that showcases feature B"],
    ["Example input 3 that showcases edge case handling"],
]

# ============================================================================
# YOUR CORE LOGIC (Replace with your actual implementation)
# ============================================================================

def process_input(user_input: str) -> str:
    """
    Your main processing function.
    
    Replace this with your actual implementation.
    """
    try:
        # Simulate processing
        time.sleep(0.5)
        
        # Your actual logic here
        result = f"Processed: {user_input}"
        
        return result
    except Exception as e:
        return f"Oops! Something went wrong. Please try a different input.\n\n(Error: {str(e)[:100]})"

# ============================================================================
# INTERFACE
# ============================================================================

# Custom theme
theme = gr.themes.Soft(
    primary_hue="indigo",
    secondary_hue="slate"
).set(
    button_primary_background_fill="*primary_500",
    button_primary_text_color="white"
)

with gr.Blocks(theme=theme, title=PROJECT_TITLE) as demo:
    # Header
    gr.Markdown(f"# {PROJECT_TITLE}")
    gr.Markdown(PROJECT_DESCRIPTION)
    
    # Main interface
    with gr.Row():
        with gr.Column():
            input_box = gr.Textbox(
                label="Input",
                placeholder="Enter your input here...",
                lines=4
            )
            submit_btn = gr.Button("Process", variant="primary")
        
        with gr.Column():
            output_box = gr.Textbox(
                label="Output",
                lines=8,
                interactive=False
            )
    
    # Examples
    gr.Markdown("### Try These Examples")
    gr.Examples(
        examples=EXAMPLES,
        inputs=[input_box],
        outputs=[output_box],
        fn=process_input,
        cache_examples=True
    )
    
    # About section
    with gr.Accordion("About This Project", open=False):
        gr.Markdown("""
        ### Technical Details
        
        [Explain the technical architecture]
        
        ### Lessons Learned
        
        [Share what you learned building this]
        
        ### Future Improvements
        
        [What would you do differently or add?]
        """)
    
    # Footer
    gr.Markdown("---")
    gr.Markdown("""
    **Contact:** [Your Name] | [GitHub](https://github.com/yourname) | [LinkedIn](https://linkedin.com/in/yourname)
    """)
    
    # Event handlers
    submit_btn.click(process_input, [input_box], [output_box])
    input_box.submit(process_input, [input_box], [output_box])

# Launch
if __name__ == "__main__":
    demo.launch()
'''

with open(f'{portfolio_dir}/gradio_template.py', 'w') as f:
    f.write(gradio_template)

print("âœ… Created Gradio portfolio template")
````

````python
# Streamlit Portfolio Template
streamlit_template = '''"""Portfolio Demo - [Your Project Name]

Author: [Your Name]
"""

import streamlit as st
import time

# ============================================================================
# PAGE CONFIG
# ============================================================================

st.set_page_config(
    page_title="[Your Project Name]",
    page_icon="ðŸš€",
    layout="wide"
)

# ============================================================================
# SIDEBAR
# ============================================================================

with st.sidebar:
    st.title("ðŸš€ [Project Name]")
    st.markdown("---")
    
    st.markdown("""
    ### About
    [Brief project description]
    
    ### Contact
    - [GitHub](https://github.com/yourname)
    - [LinkedIn](https://linkedin.com/in/yourname)
    """)
    
    st.markdown("---")
    st.caption("Built by [Your Name]")

# ============================================================================
# MAIN CONTENT
# ============================================================================

st.title("ðŸš€ [Your Project Name]")

st.markdown("""
[One-paragraph description of what your project does]

**Key Features:**
- âœ… Feature 1
- âœ… Feature 2  
- âœ… Feature 3
""")

# Quick stats
col1, col2, col3 = st.columns(3)
col1.metric("Metric 1", "Value 1")
col2.metric("Metric 2", "Value 2")
col3.metric("Metric 3", "Value 3")

st.markdown("---")

# ============================================================================
# DEMO SECTION
# ============================================================================

st.header("Try It Out")

# Example buttons
st.markdown("**Quick Examples:**")
example_cols = st.columns(3)

with example_cols[0]:
    if st.button("Example 1", use_container_width=True):
        st.session_state.demo_input = "Example input 1"

with example_cols[1]:
    if st.button("Example 2", use_container_width=True):
        st.session_state.demo_input = "Example input 2"

with example_cols[2]:
    if st.button("Example 3", use_container_width=True):
        st.session_state.demo_input = "Example input 3"

# Input/Output
user_input = st.text_area(
    "Your Input",
    value=st.session_state.get("demo_input", ""),
    placeholder="Enter your input here...",
    height=100
)

if st.button("Process", type="primary"):
    with st.spinner("Processing..."):
        time.sleep(0.5)  # Replace with actual processing
        result = f"Processed: {user_input}"
    
    st.success("Done!")
    st.markdown("### Result")
    st.write(result)

# ============================================================================
# TECHNICAL DETAILS (collapsible)
# ============================================================================

st.markdown("---")

with st.expander("ðŸ“– Technical Details"):
    st.markdown("""
    ### Architecture
    [Explain your technical approach]
    
    ### Technologies
    - Technology 1
    - Technology 2
    
    ### Challenges & Solutions
    [What problems did you solve?]
    """)
'''

with open(f'{portfolio_dir}/streamlit_template.py', 'w') as f:
    f.write(streamlit_template)

print("âœ… Created Streamlit portfolio template")
````
